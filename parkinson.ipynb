# Predicting Parkinson Disease using Machine Learning
---

## Parkinson's Disease
Parkinson's Disease (PD) is a progressive neurogenerative disorder that primarily affects motor function due to the loss of dopamine-producing neurons in the brain. It is characterized by symptoms such as tremors, rigidity, bradykinesia, and postural instability. Beyond these motor symptoms, PD also affects various non-motor functions, including cognitive, autonomic, and sensory systems. 

One significant non-motor symptom of Parkinson's Disease is its impact on voice and speech. Individuals with PD often experiance hypophonia (reduced voice volume), monotonic speech, imprecise articulation, and reduced speech rate. These vocal changes can significantly affect communication and quality of life.

## Objectives:

This notebook aims to utilize machine learning algorithms to detect the presence of the early stage PD using voice recordings of healthy people and people suffering from PD. This will serve as a preliminary project that will evolve overtime refining the model.

```python
df.drop(['name'], axis = 1, inplace = True) # we remove the name for machine learning
```

```python
X = df.drop(labels = ['status'], axis = 1)
Y = df['status']
X.head()
# we split the dataset into X and Y
```

```python
Y.head()
```

```python
# split the data further into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 40)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

---

### Machine Learning - Logistic Regression

```python
log_reg = LogisticRegression().fit(X_train, Y_train)

train_preds = log_reg.predict(X_train) # predict on train
print(f'Model accuracy on train is: {accuracy_score(Y_train, train_preds)}') # accurary on train model

test_preds = log_reg.predict(X_test) # predict on test
print(f'Model accuracy on test is: {accuracy_score(Y_test, test_preds)}')
print('-'*50)
# confusion matrix
print(f'confusion_matrix train is: {confusion_matrix(Y_train, train_preds)}')
print(f'confusion_matrix test is: {confusion_matrix(Y_test, test_preds)}')
```

---

### Machine Learning - Random Forest

```python
RF = RandomForestClassifier().fit(X_train, Y_train)

train_preds2 = RF.predict(X_train) # predict on train
print(f'Model accuracry on train is: {accuracy_score(Y_train, train_preds2)}') # accuracry on train

test_preds2 = RF.predict(X_test) # predict on test
print(f'Model accuracry on test is: {accuracy_score(Y_test, test_preds2)}') # accuracry on train
# confusion matrix
print(f'confusion_matrix train is: {confusion_matrix(Y_train, train_preds2)}')
print(f'confusion_matrix test is: {confusion_matrix(Y_test, test_preds2)}')
```
